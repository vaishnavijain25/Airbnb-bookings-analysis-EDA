# <font size="6">**Step-3 : Data Cleaning and Data Exploration**



## **Missing values**

```python
# Check count of missing values in each column
df.isnull().sum()
```

```python
# Display 'last_review' and 'reviews_per_month' columns to inspect missing or unusual values
df[['last_review','reviews_per_month']]
```

```python
# Fill missing values in 'reviews_per_month' with 0
df['reviews_per_month'].fillna(0,inplace=True)
```

```python
# Fill missing last_review dates with placeholder '25-01-2025'
df['last_review'].fillna('25-01-2025',inplace=True)
```

```python
# Remove remaining rows with missing values
df.dropna(inplace=True)
```

```python
# Display first 5 rows of the dataset
df.head()
```

```python
# Check remaining missing values after cleaning
df.isnull().sum()
```

## **Check duplicates**

```python
# Check for duplicate rows in the dataset
df.duplicated().sum()
```

## **Convert Date Columns**

```python
# Change 'last_review' column to datetime type for time-based analysis
df['last_review']=pd.to_datetime(df['last_review'],format='mixed')
```

```python
#Check data type for 'last review'
df.info()
```

## **Neighbourhood Group Analysis**

```python
# List unique values in 'neighbourhood_group' column
df['neighbourhood_group'].unique()
```

* The unique values are Manhattan, Brooklyn, Queens, Bronx, and Staten Island.

```python
# Count listings in each neighbourhood group
df['neighbourhood_group'].value_counts()
```

Observations :
* Manhattan and Brooklyn have the highest number of listings, making them the most popular boroughs on Airbnb.

* Queens, Bronx, and Staten Island have significantly fewer listings, indicating lower Airbnb activity.



## **Price Analysis by Borough**

```python
# Display first 5 price values
df['price'].head()
```

```python
# Calculate average price by neighbourhood group
df.groupby('neighbourhood_group')['price'].mean()
```

Observations:
- Manhattan has the highest average price (~$197) , indicating the most expensive borough for Airbnb listings.
- Bronx has the lowest average price (~$87), making it the most affordable.
- Brooklyn and Staten Island have mid-range average prices (~$124 and ~$115 respectively).
- Queens is slightly cheaper than Brooklyn and Staten Island (~$99).


## **Room Type Exploration**

```python
# Check unique room types available in the dataset
df['room_type'].unique()
```

```python
# Calculate average price by neighbourhood group and room type
avg_price_by_neighbourhood_group=df.groupby(['neighbourhood_group','room_type'])['price'].mean().reset_index()
avg_price_by_neighbourhood_group
```

```python
#Plot barplot for average prices for each Neighbourhood group
sns.barplot(data=avg_price_by_neighbourhood_group,x='neighbourhood_group',y='price',hue='room_type',palette='Set2')
plt.show()
```

Observations:
* Manhattan has the highest average prices across all room types, especially for entire homes (~$249).

* Brooklyn and Staten Island follow, while Bronx and Queens offer more budget-friendly options.

* Shared rooms are the most affordable, especially in Brooklyn and Staten Island.

## **Review Date and Temporal Analysis**

```python
# Sort review dates to identify oldest and most recent reviews
df['last_review'].sort_values()
```

```python
# Extract year from 'last_review' for temporal analysis
df['year']=df['last_review'].dt.year
```

```python
df
```

```python
# Exclude rows with year 2025 (placeholder) from dataset
df1=df[df['year']!=2025]
df1
```

## **Listings Over the Years**

```python
# Count listings per year and convert to DataFrame
year_vs_listing=df1.groupby('year')['id'].count().reset_index()
year_vs_listing
```

```python
# Plot yearly listing counts using a bar chart
sns.barplot(data=year_vs_listing,x='year',y='id')
plt.show()
```

Observations:

* Listings increased steadily from 2011 to 2019.

* Sharp growth after 2015, peaking in 2019.

* Indicates rising popularity and market expansion over time.

## **Listings Count Over Years by Neighbourhood Group**

```python
# Group by Neighbourhood group and year to count hosts
year_vs_neighbor=df1.groupby(['neighbourhood_group','year'])['host_id'].count().reset_index()
# Rename column
year_vs_neighbor.rename(columns={'host_id':'count'},inplace=True)
```

```python
year_vs_neighbor
```

```python
# Bar plot showing yearly listing counts by borough
sns.barplot(data=year_vs_neighbor, x='year', y='count', hue='neighbourhood_group')
```

Observations:

* Brooklyn and Manhattan show the highest host activity, especially in 2019, with over 10,000 listings each — indicating these are the most active boroughs.

* Queens has moderate growth, reaching over 3,000 listings by 2019.

* Bronx shows a gradual increase, reaching nearly 700 listings in 2019.

* Staten Island has the least number of listings throughout, indicating limited Airbnb activity.

* There’s a clear upward trend in all boroughs over time, highlighting the growing popularity of Airbnb in NYC.

## **Number of Reviews for each Neighbourhood Group using PieChart**

```python
# Count total number of reviews for each neighbourhood group
total_reviews_by_neighbourhood_group= df.groupby('neighbourhood_group')['number_of_reviews'].count()
total_reviews_by_neighbourhood_group
```

```python
 # Plot a pie chart showing percentage of total reviews by neighbourhood group
 plt.pie(total_reviews_by_neighbourhood_group,labels=total_reviews_by_neighbourhood_group.index,autopct='%1.1f%%')
 plt.title("Number of reviews by Neighbourhood Groups",fontsize=15)
 plt.show()
```

**Observations:**
* Manhattan has the highest number of reviews (44.3%), indicating strong guest engagement and popularity.

* Brooklyn follows closely with 41.1%, also showing high activity and interest.

* Queens (11.6%) and Bronx (2.2%) have significantly fewer reviews, suggesting moderate guest engagement.

* Staten Island has the lowest number of reviews (0.8%), reflecting minimal listing presence or visitor activity.
